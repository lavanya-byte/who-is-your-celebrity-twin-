```python
from keras_vggface.utils import preprocess_input
from keras_vggface import VGGFace
import numpy as np
import pickle
from sklearn.metrics.pairwise import cosine_similarity
import cv2
from mtcnn import MTCNN
from PIL import Image
```

```python
feature_list=np.array(pickle.load(open('embedding.pkl','rb')))
```

```python
filesnames=pickle.load(open('filenames.pkl','rb'))
```

```python
model=VGGFace(model='resnet50',include_top=False,input_shape=(224,224,3),pooling='avg')
```

```python
detector=MTCNN()
```

```python
sample_img=cv2.imread('sample/little.jpg')
```

```python
result=detector.detect_faces(sample_img)
```

```python
x,y,width,height=result[0]['box']
face=sample_img[y:y+height,x:x+width]
```

```python
image=Image.fromarray(face)
image=image.resize((224,224))
face_array=np.asarray(image)
face_array=face_array.astype('float32')
expanded_img=np.expand_dims(face_array,axis=0) 
# expand dimensions
preprocessed_img=preprocess_input(expanded_img)
result=model.predict(preprocessed_img).flatten()
# print(result)
# print(result.shape)
```

```python
# print(cosine_similarity(result.reshape(1,-1),feature_list[0].reshape(1,-1))[0][0])
```

```python
similarity=[]
for i in range(len(feature_list)):
    similarity.append(cosine_similarity(result.reshape(1,-1),feature_list[i].reshape(1,-1))[0][0])
# print(len(similarity))    
```

```python
print(list(enumerate(similarity)))
# basically hum indexes add krre h similarity ko index dere h and then we will sort on the basis of second item basis
```

```python
sorted(list(enumerate(similarity)),reverse=True,key=lambda x:x[1])
# basically humne chota sa code likha h which is telling that we have to sort for a secondd item of the list
```

```python
index_pos=sorted(list(enumerate(similarity)),reverse=True,key=lambda x:x[1])[0][0]
```

```python
temp=cv2.imread(filesnames[index_pos])
cv2.imshow('output',temp)
cv2.waitKey(0)
```
