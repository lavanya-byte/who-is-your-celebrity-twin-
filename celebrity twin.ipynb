```python
import os
import pickle
```

```python
actors=os.listdir('data')
```

```python
print(actors)
```

```python
filenames=[]
```

```python
for actor in actors:
    for file in os.listdir(os.path.join('data',actor)):
        filenames.append(os.path.join('data',actor,file))
# copying the path of each and every actor image and will be used for feature extraction
```

```python
print(len(filenames))
```

```python
pickle.dump(filenames,open('filenames.pkl','wb'))
# dumping in write binary mode
```

```python
from tensorflow.keras.preprocessing import image
from keras_vggface.utils import preprocess_input
from keras_vggface.vggface import VGGFace
import numpy as np
```

```python
filenames=pickle.load(open('filenames.pkl','rb'))
```

```python
model=VGGFace(model='resnet50',include_top=False,input_shape=(224,224,3),pooling='avg')
print(model.summary())
# whole synatax of VGGFace
```

```python
def fr(img_path,model):
    img=image.load_img(img_path,target_size=(224,224))
    img_array=image.img_to_array(img)
    expanded_img=np.expand_dims(img_array,axis=0)
    preprocessed_img=preprocess_input(expanded_img)
    result=model.predict(preprocessed_img).flatten()
    return result
# main function
```

```python
features=[]
from tqdm import tqdm
```

```python
for file in tqdm(filenames):
    features.append(fr(file,model))
```

```python
pickle.dump(features,open('embedding.pkl','wb'))
```
